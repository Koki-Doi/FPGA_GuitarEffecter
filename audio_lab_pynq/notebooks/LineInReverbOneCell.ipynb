```python
# Line-in -> adjustable reverb -> output jack, one-cell setup
import audio_lab_pynq as aud
from IPython.display import display

ol = aud.AudioLabOverlay()

def configure_output_path():
    # Input path: Line-in AUX -> ADC
    ol.codec.R4_RECORD_MIXER_LEFT_CONTROL_0 = 0x01
    ol.codec.R5_RECORD_MIXER_LEFT_CONTROL_1 = 0x05
    ol.codec.R6_RECORD_MIXER_RIGHT_CONTROL_0 = 0x01
    ol.codec.R7_RECORD_MIXER_RIGHT_CONTROL_1 = 0x05
    ol.codec.R19_ADC_CONTROL = 0x03

    # Digital playback path: serial input -> DAC -> playback mixers
    ol.codec.R58_SERIAL_INPUT_ROUTE_CONTROL = 0x01
    ol.codec.R59_SERIAL_OUTPUT_ROUTE_CONTROL = 0x01
    ol.codec.R36_DAC_CONTROL_0 = 0x03
    ol.codec.R22_PLAYBACK_MIXER_LEFT_CONTROL_0 = 0x21
    ol.codec.R24_PLAYBACK_MIXER_RIGHT_CONTROL_0 = 0x41

    # PYNQ-Z2 audible jack path: playback mixers -> LOUT/ROUT
    ol.codec.R26_PLAYBACK_LR_MIXER_LEFT_LINE_OUTPUT_CONTROL = 0x03
    ol.codec.R27_PLAYBACK_LR_MIXER_RIGHT_LINE_OUTPUT_CONTROL = 0x09
    ol.codec.R31_PLAYBACK_LINE_OUTPUT_LEFT_VOLUME_CONTROL = 0xE7
    ol.codec.R32_PLAYBACK_LINE_OUTPUT_RIGHT_VOLUME_CONTROL = 0xE7

    # Keep headphone path enabled too; harmless if the jack is wired to LOUT/ROUT.
    ol.codec.R29_PLAYBACK_HEADPHONE_LEFT_VOLUME_CONTROL = 0xE7
    ol.codec.R30_PLAYBACK_HEADPHONE_RIGHT_VOLUME_CONTROL = 0xE7
    ol.codec.R35_PLAYBACK_POWER_MANAGEMENT = 0x03

    # Codec clocks/DSP
    ol.codec.R61_DSP_ENABLE = 0x01
    ol.codec.R62_DSP_RUN = 0x01
    ol.codec.R65_CLOCK_ENABLE_0 = 0x7F
    ol.codec.R66_CLOCK_ENABLE_1 = 0x03

configure_output_path()

def apply_reverb(enabled=True, reverb=35, tone=70, mix=25):
    word = ol.set_reverb(enabled=enabled, reverb=reverb, tone=tone, mix=mix)
    route = "line_in -> reverb -> output jack" if enabled else "line_in -> passthrough -> output jack"
    print(route)
    print("enabled:", enabled, "reverb:", reverb, "tone:", tone, "mix:", mix, "control:", hex(word))
    return word

try:
    import ipywidgets as widgets

    enabled_widget = widgets.Checkbox(value=True, description="Reverb")
    reverb_widget = widgets.IntSlider(value=35, min=0, max=100, step=1, description="reverb", continuous_update=False)
    tone_widget = widgets.IntSlider(value=70, min=0, max=100, step=1, description="tone", continuous_update=False)
    mix_widget = widgets.IntSlider(value=25, min=0, max=100, step=1, description="mix", continuous_update=False)
    output = widgets.Output()

    def update(_=None):
        with output:
            output.clear_output(wait=True)
            apply_reverb(
                enabled=enabled_widget.value,
                reverb=reverb_widget.value,
                tone=tone_widget.value,
                mix=mix_widget.value,
            )

    for widget in [enabled_widget, reverb_widget, tone_widget, mix_widget]:
        widget.observe(update, names="value")

    display(widgets.VBox([enabled_widget, reverb_widget, tone_widget, mix_widget, output]))
    update()
except Exception as exc:
    print("ipywidgets is not available; use apply_reverb(...) manually.")
    print("Example: apply_reverb(enabled=True, reverb=35, tone=70, mix=25)")
    print("Widget error:", exc)
    apply_reverb(enabled=True, reverb=35, tone=70, mix=25)
```
